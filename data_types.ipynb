```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Creating a sample dataset with different data types
data = {
    'age': [22, 25, 30, 35, 40, 45, 50, 55],
    'salary': [40000, 50000, 60000, 65000, 70000, 80000, 90000, 100000],
    'gender': ['Male', 'Female', 'Female', 'Male', 'Male', 'Female', 'Male', 'Female'],
    'purchased': [True, False, True, False, True, False, True, False],
    'date_of_birth': pd.to_datetime(['1998-04-12', '1995-06-21', '1990-02-14', '1985-11-05', '1980-07-19', '1975-10-11', '1970-03-03', '1965-09-28'])
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Display the first few rows
print(df)
```

## Numerical Data

Numerical data like age and salary can be visualized using histograms, box plots, or scatter plots.

```python
# Numerical Data: Age Histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=8, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Numerical Data: Scatter Plot of Age vs. Salary
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='salary', data=df)
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

# Categorical Data
Categorical data like gender can be visualized using count plots or bar plots.

```python
# Categorical Data: Count Plot for Gender
plt.figure(figsize=(8, 5))
sns.countplot(x='gender', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

# Boolean Data

Boolean data like purchased can also be visualized using count plots.

```python
# Boolean Data: Count Plot for Purchased
plt.figure(figsize=(8, 5))
sns.countplot(x='purchased', data=df)
plt.title('Purchased Status Distribution')
plt.xlabel('Purchased')
plt.ylabel('Count')
plt.show()
```

# Date/Time Data

Date/time data like date_of_birth can be processed and visualized to identify trends over time.

```python
# Date Data: Line Plot of Age over Time (Date of Birth)
df['year'] = df['date_of_birth'].dt.year

# Plot Salary by Year of Birth
plt.figure(figsize=(10, 6))
sns.lineplot(x='year', y='salary', data=df)
plt.title('Salary vs Year of Birth')
plt.xlabel('Year of Birth')
plt.ylabel('Salary')
plt.show()
```
