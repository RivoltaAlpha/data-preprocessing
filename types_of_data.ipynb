```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Creating a sample dataset
data = {
    'age': [25, 30, 45, 35, 50, 40, 60, 55],
    'salary': [50000, 60000, 80000, 75000, 120000, 95000, 110000, 105000],
    'gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'purchased': ['No', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No'],
    'education_level': ['High School', 'College', 'College', 'University', 'High School', 'University', 'University', 'College']
}

# Create DataFrame
df = pd.DataFrame(data)

# Display the first few rows
print(df)
```

#   Visualizing Different Types of Data

# Categorical

```python
# Continuous Data: Histogram of Age
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Continuous Data: Scatter Plot of Age vs. Salary
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='salary', data=df, hue='gender')
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

# Continous

```python
# Categorical Data: Countplot for Gender
plt.figure(figsize=(8, 5))
sns.countplot(x='gender', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Categorical Data: Countplot for Purchased
plt.figure(figsize=(8, 5))
sns.countplot(x='purchased', data=df)
plt.title('Purchased Distribution')
plt.xlabel('Purchased')
plt.ylabel('Count')
plt.show()
```

# Ordinal

```python
# Convert 'education_level' to a categorical type with an order
education_order = ['High School', 'College', 'University']
df['education_level'] = pd.Categorical(df['education_level'], categories=education_order, ordered=True)

# Ordinal Data: Countplot for Education Level
plt.figure(figsize=(10, 6))
sns.countplot(x='education_level', data=df, order=education_order)
plt.title('Education Level Distribution')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.show()
```

When to Use Certain Visualizations
## Data Visualization Types

### Continuous Data

- **Histograms**, **scatter plots**, and **box plots** are ideal for showing the distribution or relationship between variables.

### Categorical Data

- **Bar plots** or **count plots** are perfect for showing category distribution.

### Ordinal Data

- **Ordered bar plots** or **line plots** help to keep the natural order intact while visualizing.



Visualizing continuous, categorical, and ordinal data helps in understanding their distributions and relationships. 

This understanding plays a crucial role in preprocessing data before feeding it into machine learning models.
