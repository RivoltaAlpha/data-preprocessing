```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Sample dataset
data = {
    'gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'color': ['Red', 'Blue', 'Green', 'Red', 'Blue'],
    'salary': [50000, 60000, 55000, 70000, 65000]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Display dataset
print(df)

```

 ### Label Encoding (for gender)

```python
# Initialize Label Encoder
le = LabelEncoder()

# Apply Label Encoding to the 'gender' column
df['gender_encoded'] = le.fit_transform(df['gender'])

# Display dataset with encoded gender column
print(df)
```

### One-Hot Encoding (for color)

One-Hot Encoding creates binary (0 or 1) columns for each category in a nominal feature. 

Each row will have a 1 in the column corresponding to the category and 0 in others.

```python
# Apply One-Hot Encoding to the 'color' column
df = pd.get_dummies(df, columns=['color'], drop_first=True)

# Display dataset with One-Hot Encoded columns
print(df)
```

## Visualizing the Encoded Data

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualizing salary by color (after encoding)
sns.boxplot(x='color_Red', y='salary', data=df)
plt.title('Salary Distribution by Color (Red vs Others)')
plt.show()
```
